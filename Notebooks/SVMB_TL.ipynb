```python
def plotRealVel(Starting):
    
    #create figure
    fig = plt.figure(figsize=(15,15))
    
    rows = 10
    columns = 10
    
    for i in range(100):
        fig.add_subplot(rows,columns,i+1)
        #kk=starting+i
        #filename='/disk/student/adhara/Fall2021/FCNVMB-Deep-learning-based-seismic-velocity-model-building/Output/'+str(i+1)+'_real_BT.npy'
        mat=file[Starting+i,:,:]
        mat = mat*1000
        fig.add_subplot(rows, columns, i+1)
        plt.imshow(np.squeeze(mat))
        plt.axis('off')
        plt.tight_layout()
```

```python
plotRealVel(500)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import test_scalar
import numpy as np
```

## Direct

```python
y = test_scalar.model_direct_2d(propagator=test_scalar.scalarprop)

fig, ax = plt.subplots(2,2, figsize=(8,8))
for shot in range(2):
    for receiver in range(2):
        ax[shot][receiver].plot(y[0][:, shot, receiver].numpy())
        ax[shot][receiver].plot(y[1][:, shot, receiver].numpy())
diff = (y[0] - y[1]).numpy().ravel()
np.linalg.norm(diff)
```

## Scatter

```python
y = test_scalar.model_scatter_2d(propagator=test_scalar.scalarprop,
                                 prop_kwargs={'pml_width': 30})

fig, ax = plt.subplots(2,2, figsize=(8,8))
for shot in range(2):
    for receiver in range(2):
        ax[shot][receiver].plot(y[0][:, shot, receiver].numpy())
        ax[shot][receiver].plot(y[1][:, shot, receiver].numpy())
diff = (y[0] - y[1]).numpy().ravel()
np.linalg.norm(diff)
```

## Grad

```python
y = test_scalar.model_grad_2d(propagator=test_scalar.scalarprop,
                              prop_kwargs={'pml_width': 30})

vmin, vmax = np.nanpercentile(y[0].numpy(), [2,98])
plt.imshow(y[0][0].numpy(), vmin=vmin, vmax=vmax)
plt.figure()
plt.imshow(y[1][0].numpy(), vmin=vmin, vmax=vmax)
diff = (y[0]-y[1]).numpy().ravel()
np.linalg.norm(diff[np.where(~np.isnan(diff))])
```
